```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train=pd.read_csv(r"C:\Users\Sunil D Kaushik\Downloads\train.csv")
test=pd.read_csv(r"C:\Users\Sunil D Kaushik\Downloads\test (1).csv")
train["total"]=train["RoomService"]+train["FoodCourt"]+train["ShoppingMall"]+train["Spa"]+train["VRDeck"]
test["total"]=test["RoomService"]+test["FoodCourt"]+test["ShoppingMall"]+test["Spa"]+test["VRDeck"]
train.drop(['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck','Name'],axis=1,inplace=True)
test.drop(['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck','Name'],axis=1,inplace=True)
train[['GroupNum','Group_Id']] =train['PassengerId'].str.split('_', expand=True)

train[['deck','num','side']] = train['Cabin'].str.split('/', expand=True)
test[['GroupNum','Group_Id']] =test['PassengerId'].str.split('_', expand=True)

test[['deck','num','side']] = test['Cabin'].str.split('/', expand=True)
train.drop(['PassengerId','Cabin'],axis=1,inplace=True)
test.drop(['PassengerId','Cabin'],axis=1,inplace=True)
from sklearn.impute import SimpleImputer


doc = {} 
def imputer_func(col, data, strategy='mean'):
    imp = SimpleImputer(strategy= strategy)
    data[col] = imp.fit_transform(data[[col]])
    doc[col] = imp
    return data[col]

train['HomePlanet'] = imputer_func('HomePlanet',train,strategy = 'most_frequent')
train['CryoSleep'] = imputer_func('CryoSleep',train,strategy = 'most_frequent')
train['Destination'] = imputer_func('Destination',train,strategy = 'most_frequent')
train['Age'] = imputer_func('Age',train,strategy = 'mean')
train['VIP'] = imputer_func('VIP',train,strategy = 'most_frequent')
train['total'] = imputer_func('total',train,strategy = 'mean')
train['deck'] = imputer_func('deck',train,strategy = 'most_frequent')
train['num'] = imputer_func('num',train,strategy = 'mean')
train['side'] = imputer_func('side',train,strategy = 'most_frequent')
train['Transportede']=train['Transported']
train.drop(['Transported'],axis=1,inplace=True)
train['Transported']=train['Transportede']
train.drop(['Transportede'],axis=1,inplace=True)
def f(s) :
    if s=="Earth" :
     return 0
    if s=="Europa" :
     return 1
    if s=="Mars" :
     return 2
train['HomePlanet']=train.HomePlanet.apply(f)
train
```
